## Conclusiones

Con todo esto, podemos responder a las preguntas de la siguiente manera:

1) ¿Dónde estuvo el mayor consumo de tiempo?

En este caso, tuvimos un mayor tiempo haciendo uso de los datos remotos de Hugging Face. Este comportamiento se explica porque antes de ejecutar la operación de suma fue necesario transferir el conjunto de datos a través de la red, lo que introduce latencia de comunicación, sobrecarga de protocolos y dependencia del ancho de banda disponible. En este caso, el tiempo de transferencia supera ampliamente el tiempo de cómputo requerido para realizar la operación matemática.

2) ¿Qué escenario minimiza la transferencia de datos?

El escenario que minimiza la transferencia de datos es aquel en el que los datos se almacenan y procesan localmente en la misma máquina. En esta configuración no existe transmisión de información por red, lo que elimina completamente la latencia asociada al movimiento de datos y permite que el tiempo de ejecución dependa únicamente del acceso a memoria o disco local y del procesamiento del CPU. Por esta razón, este escenario representa la situación más eficiente cuando el volumen de datos y el procesamiento se concentran en un único nodo.

3)  ¿Cuál maximiza la latencia?

El escenario que maximiza la latencia corresponde al uso de datos remotos procesados localmente, ya que la ejecución depende de la transferencia de información a través de redes externas antes de iniciar el procesamiento. La latencia surge principalmente del tiempo de establecimiento de conexiones, la transmisión de paquetes y las posibles variaciones en el rendimiento de la red. Este retardo ocurre independientemente de la simplicidad del cálculo realizado, por lo que incluso operaciones computacionalmente ligeras pueden experimentar incrementos significativos en el tiempo total debido a la comunicación remota.

4) ¿Qué arquitectura escala mejor?

La arquitectura que mejor escala es aquella basada en el uso de almacenamiento en bucket dentro de un entorno cloud junto con máquinas virtuales en la misma infraestructura, como el escenario implementado en GCP. Esta configuración permite aumentar recursos de cómputo de manera vertical (mayor capacidad de CPU y memoria) o horizontal (más instancias en paralelo), además de facilitar el uso de servicios distribuidos. Gracias a estas características, el sistema puede adaptarse a volúmenes de datos mucho mayores sin depender de las limitaciones físicas de una sola máquina local.

5) ¿Qué porcentaje del tiempo total se invierte en mover datos?

Este cálculo podemos hacerlo de la siguiente manera:

$[ (tiempohf - tiempolocal) / (tiempohf) ] * 100$

Y con esta comparación, se puede estimar que aproximadamente entre el 70% y el 75% del tiempo total de ejecución en el caso remoto se dedica al movimiento de datos y no al procesamiento. Esta estimación se obtiene al considerar la diferencia entre ambos tiempos, asumiendo que el costo computacional de la operación de suma es prácticamente el mismo en los dos casos. El resultado evidencia que, en tareas de bajo costo computacional, la transferencia de datos suele convertirse en el factor dominante del tiempo total de ejecución.
